```python
def column_explore (x):
    print("Column Number of rows: ",len(x))
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
    print("Number of Unique values:",x.nunique())
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
    print(f"% of nan values is: ",round(100*(x.isnull().sum() / len(x)),2),"%")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
    print("Column Data Type: ",x.dtype)
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
    print("COLUMN UNIQUE VALUES: \n \n",x.unique())
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
    print("VARIABLE WEIGHT WITHIN THE TOTAL OF THE COLUMN: \n \n",round((x.value_counts() / x.value_counts().sum())*100,2))
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ")
```

```python
column_explore(df["Injury"])
```
