# We drop the columns that we know we are not going to use

```python
df1 = df.drop(columns = ['original order','case number_2','case number_1','href','href formula','pdf',
                                'investigator or source','unnamed: 22','unnamed: 23','name',"new_index",])
df1.head()
```

# We drop the duplicates

```python
df1.drop_duplicates(inplace=True) 
```

```python
df1.columns = ['Date', 'Year', 'Type', 'Country', 'Area','Location','Activity',
                   'Sex', 'Age','Injury', 'Fatal', 'Time', 'Species']
df1.head()
```

We drop some more unnecessary columns for our research

```python
df1.drop(columns =["Time","Date","Type","Location","Activity"])
```

# We drop all null values

# We want the "sex" column to be either Male or Female so we apply the following function

```python
df1['Sex'] = df1['Sex'].str.strip()
df1['Sex'].replace(['N','.','lli'],np.nan,inplace=True)
df1['Sex'].replace(['M'], 'Male', inplace=True)
df1['Sex'].replace(['F'], 'Female', inplace=True)
df1['Sex'].value_counts() 
```

# We want our "age" column to be integers and if its not possible change to Na

```python
df1['Age'] = pd.to_numeric(df1['Age'], errors='coerce')
df1["Age"].value_counts()
```
